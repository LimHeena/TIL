```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
```

```python
call = pd.read_csv('call_data_scaled.csv')
call_data = pd.read_csv('call_data.csv')
```

### 해지여부를 예측하는 모델링
- 분류
- 0,1 이진분류
- 이진분류를 하는 모델링
- LGBM, XGM, CAT, RF 등 다양하게 분류에 대한 알고리즘을 사용
- y값은 전화해지여부
- x값은 나머지

```python
call = call.drop(['Unnamed: 0'], axis=1)
call
```

```python
y = call_data[['전화해지여부']]
result = pd.concat([call,y], axis=1)
result
```

```python
result.columns
```

- 데이터를 학습하기 위해서 데이터를 나눠서 학습시켜야 한다.
- train, test split 진행
- 교차검증 cross_validate
- 교차검증이란, split 랜덤으로 나눠서 kfold, Skfold

```python
df_train = result[['가입일', '음성사서함이용', '상담전화건수', '주간통화시간', '저녁통화시간', '밤통화시간',
       '주간통화횟수', '저녁통화횟수', '밤통화횟수', '주간통화요금', '저녁통화요금', '밤통화요금']]
df_y = result['전화해지여부']
```

```python
df_train
```

```python
df_y
```

```python
model = [DecisionTreeClassifier(), RandomForestClassifier(), LogisticRegression()]
name = ['DT','RF','LR']

for model, name in zip(model, name):
    print('########## 사용하는 알고리즘',name,'########')
    for score in ['accuracy','precision','recall','f1']:
        print(score)
        print('--')
        print(cross_val_score(model, df_train, df_y, scoring=score, cv=3 ))
```
