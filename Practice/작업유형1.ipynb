# 첫번째 문제

```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('car_crashes')
print(df.head())
```

```python
data = df.sort_values(by=df.columns[2], ascending=False)
data.head()
```

```python
data.iloc[9,2]
```

```python
ten = data.iloc[9,2]
data.iloc[:10,2] = ten
data.head(11)
```

```python
result = data[(data['speeding'] >= 7)]
print('최종 결과물은 ', result['alcohol'].mean())
```

# 두번째 문제

```python
import seaborn as sns
import pandas as pd
import numpy as np

df = sns.load_dataset('planets')
df.head()
```

```python
len(df)
```

```python
len(df)*0.7
```

```python
int(len(df)*0.7)
```

```python
len = int(len(df)*0.7)
df = df[:len]
df2 = df.copy()
df2['distance'] = df2['distance'].fillna(df2['distance'].median())
```

```python
print('전처리 전: ', np.std(df['distance']))
print('전처리 후: ', np.std(df2['distance']))
```

# 세번째 문제

```python
import seaborn as sns

df = sns.load_dataset('planets')
df.head()
```

```python
df['orbital_period'].quantile(0.25)
```

```python
df['orbital_period'].quantile(0.75)
```

```python
Q25 = df['orbital_period'].quantile(0.25)
Q75 = df['orbital_period'].quantile(0.75)
IQR = Q75-Q25
min = Q25 - IQR*1.5
max = Q75 + IQR*1.5
```

```python
print(IQR)
print(min)
print(max)
```

```python
df_outlier = df[(df['orbital_period'] <= min) | (df['orbital_period'] >= max)]
df_outlier
```

```python
print(df_outlier['orbital_period'].sum())
```
