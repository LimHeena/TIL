```python
# 이미지 하나만 저장 

import requests
import os

def download_image(url, save_path):
    response = requests.get(url, stream=True)
    if response.status_code == 200:
        with open(save_path, 'wb') as file:
            for chunk in response.iter_content(1024):
                file.write(chunk)
        print("이미지 다운로드 완료:", save_path)
    else:
        print("이미지 다운로드 실패")

# 이미지 URL과 저장할 파일 경로 설정
image_url = "https://shopping-phinf.pstatic.net/main_3854436/38544367620.20230310142318.jpg?type=f640"
save_location = "/content/image/image1.jpg"  # 다운로드한 이미지를 저장할 경로와 파일명

# 이미지 다운로드 실행
download_image(image_url, save_location)
```

```python
# 이미지 여러개 저장해서 zip으로 생성 

import requests
import zipfile
import os

def download_image(url, save_path):
    response = requests.get(url, stream=True)
    if response.status_code == 200:
        with open(save_path, 'wb') as file:
            for chunk in response.iter_content(1024):
                file.write(chunk)
        print("이미지 다운로드 완료:", save_path)
        return True
    else:
        print("이미지 다운로드 실패")
        return False

def create_zip(directory, zip_name):
    zip_path = f"{zip_name}.zip"
    with zipfile.ZipFile(zip_path, 'w') as zipf:
        for root, _, files in os.walk(directory):
            for file in files:
                file_path = os.path.join(root, file)
                zipf.write(file_path, os.path.relpath(file_path, directory))
    print("ZIP 파일 생성 완료:", zip_path)

# 이미지 URL 리스트
image_urls = [
    "https://shopping-phinf.pstatic.net/main_3854436/38544367620.20230310142318.jpg?type=f640",
    "https://shopping-phinf.pstatic.net/main_4002140/40021405295.20230514115806.jpg?type=f640",
    "https://shopping-phinf.pstatic.net/main_3877844/38778449622.20230320112916.jpg?type=f640"
]

# 이미지를 저장할 디렉토리
save_directory = "/content/images"

# 이미지 다운로드 및 저장
for i, url in enumerate(image_urls):
    image_name = f"image{i+1}.jpg"
    save_path = os.path.join(save_directory, image_name)
    download_image(url, save_path)

# 다운로드된 이미지들을 하나의 ZIP 파일로 묶기
create_zip(save_directory, "naver_images")
```
